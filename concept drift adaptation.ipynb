```python
win_size=40
a=0.91
b=0.88
count_WL=0

yt=[]
yp=[]
xt=[]
d=0
i=0
x_new=[]
y_new=[]
counter_wl=0
counter_dl=0
X=train_df.drop(['treatment'],axis=1)
y=train_df.treatment

for xi,yi in stream.iter_pandas(X,y):
    xi2=np.array(list(xi.values()))
    y_pred=classifier.predict(xi2.reshape(1, -1))      # make a prediction
    #print(i)
    yt.append(yi)
    yp.append(y_pred)
    xt.append(xi2)
    #print(yt,yp,xt)
   
    #metric=metric.update(yi, y_pred[0])
    if i>=2*win_size and (i)%win_size==0:
        print("Evaluating")
        acc_currWin=accuracy_score(yt[i-win_size:],yp[i-win_size:])
        acc_lastWin=accuracy_score(yt[i-2*win_size:i-win_size],yp[i-2*win_size:i-win_size])
        #print(i-win_size)
        print(acc_currWin,acc_lastWin)
        
        if((acc_lastWin-acc_currWin)>0.05):
            print("Difference in accuracy",abs(acc_currWin-acc_lastWin))
    

        curr_win=(int)(i/win_size)
        print(curr_win)
        
        if((acc_currWin<a*acc_lastWin) and count_WL==0):
            print("Drift Detected(Warning Level) at window no. ",curr_win)
            counter_wl+=1
            count_WL=1
        elif(count_WL==1 and (acc_currWin<a*acc_lastWin) and (acc_currWin<b*acc_lastWin)):
            print("Drift Detected(Drift Level) at window no. ",curr_win)
            #print(i-2*win_size,i)
            counter_dl+=1
            classifier.fit(xt[i-2*win_size:],yt[i-2*win_size:]) 
        else :
            count_WL=0
    i+=1
           
print("warning level detected ",counter_wl," times")
print("drift level detected ",counter_dl," times")
```
